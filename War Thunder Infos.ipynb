```python
# get csv loaded
import pandas as pd
df = pd.read_csv('WT Help - Success Rates.csv')

# if I didn't record the country I didn't record
df.dropna(subset=['Country'], inplace=True)
df

```

```python
from IPython.display import display

# most number of kills
df['all kills'].max()

# average number of kills
df['all kills'].mean()

# average number of deaths
df['Deaths'].mean()

basic_df = df.groupby(['Country', 'Vehicle', 'BR']).agg({
    'all kills': ['mean', 'max'],
    'Assists ': ['mean', 'max'],
    'Deaths': ['mean', 'max']}).reset_index().round(2)

basic_df['ka-d ratio'] = (basic_df['all kills']['mean'] + basic_df['Assists ']['mean']) / basic_df['Deaths']['mean']

with pd.option_context('display.max_rows', 500):
    #display(basic_df['all kills']['mean'])
    display(basic_df.sort_values(by=['ka-d ratio'], ascending=False))

```
